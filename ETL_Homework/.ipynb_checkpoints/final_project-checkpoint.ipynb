```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
#import tattoo csv
csv_file = "nba-tattoos-data.csv"
tattoos_df = pd.read_csv(csv_file)
tattoos_df.head()
```

```python
tattoos_df = tattoos_df.rename(index=str, columns={"Player Name": "playerName", "Tattoos yes/no": "tattooed"})
tattoos_df.head()
```

```python
#import stats csv
csv_file = "nba-stats-data.csv"
stats_df = pd.read_csv(csv_file)
stats_df.head()
```

```python
#drop Rk from table
stats_df = stats_df.drop(['Rk'], axis=1)
```

```python
#connect to db
conn_string = "mysql://root:Depreston5@localhost:3306/nba_db"
engine = create_engine(conn_string)
```

```python
#load dfs to sql
stats_df.to_sql(name='stats', con=engine, if_exists='replace', index=False)
tattoos_df.to_sql(name='tattoo', con=engine, if_exists='replace', index=False)
```

```python
#confirm sql db has been written
pd.read_sql_query('select * from stats', con=engine).head()
```

```python
#get total points and players for tattooed and nontattooed players, had to drop null games played from tattoo list since more players were listed than played in the season being analyzed
pd.read_sql_query('SELECT t.tattooed, sum(s.PTS) as points, count(t.playerName) as players FROM tattoo t Left JOIN stats s ON t.playerName=s.Player WHERE s.G IS NOT NULL GROUP BY t.tattooed;', con=engine).head()
```
